```python
import numpy as np
import matplotlib.pyplot as plt

n_x = np.arange(1,10.,1)
n_y = np.arange(1,10.,1)

X,Y = np.meshgrid(n_x,n_y)

R = np.sqrt(X**2+Y**2)

k = 6
dk = 0.5

x_r1 = np.arange(0,k,0.01)
y_r1 = np.sqrt(k**2-x_r1**2)

x_r2 = np.arange(0,k+dk,0.01)
y_r2 = np.sqrt((k+dk)**2-x_r2**2)

R = np.sqrt(X**2+Y**2)
R_s = R*((R>k).astype(int))
R_s = R_s*((R<(k+dk)).astype(int))

fig,ax = plt.subplots(1,1,figsize=[5,5],dpi=150)
ax.scatter(X,Y)
ax.scatter(X[R_s>0],Y[R_s>0],color='C1')
ax.plot(x_r1,y_r1,'C1')
ax.plot(x_r2,y_r2,'C1')
ax.arrow(0,0,4.23,4.23)
ax.arrow(4.23,4.23,0.38,0.38,color='C1')
ax.text(2.5,2.9,"k")
ax.text(4.1,4.5,"dk",color='C1')
ax.set_xticks(n_x)
ax.set_xticklabels([r"$\pi/\hbar$",r"$2\pi/\hbar$",r"$3\pi/\hbar$",r"$4\pi/\hbar$",r"$5\pi/\hbar$",r"$6\pi/\hbar$",r"$7\pi/\hbar$",r"$8\pi/\hbar$",r"$9\pi/\hbar$"])
ax.set_yticks(n_x)
ax.set_yticklabels([r"$\pi/\hbar$",r"$2\pi/\hbar$",r"$3\pi/\hbar$",r"$4\pi/\hbar$",r"$5\pi/\hbar$",r"$6\pi/\hbar$",r"$7\pi/\hbar$",r"$8\pi/\hbar$",r"$9\pi/\hbar$"])
ax.set_xlim(0.5,9.5)
ax.set_xlabel(r"$k_{x}$")
ax.set_ylabel(r"$k_{y}$")
ax.set_title(r"k-space for $k_z=\pi/\hbar$ ($n_z=1$).")
ax.set_ylim(0.5,9.5)
plt.savefig("../Figures/k_space_2D.png")
plt.show()
```
