```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.1) # Just some arbitrary range to calculate x over.
a = 1
C = 1/(2*np.pi*a**2)
P = C*np.exp(-x**2/a**2) # I've centered this at <x>=0.

fig = plt.figure(figsize=[4,2.5],dpi=150)
ax = fig.add_subplot(111)
ax.plot(x,P,'k-')
ax.axvline(-a,linestyle='--',color='C1')
ax.axvline(0,linestyle='-.',color='C0')
ax.axvline(a,linestyle='--',color='C1')

ax.set_xlabel("x")
ax.set_ylabel("P(x)")
ax.set_xlim(min(x),max(x))
ax.set_ylim(0,)
ax.set_xticks([-1,0,1],labels=['-a',r'$\left< {\rm x} \right>$','a']) #Setting custom x ticks
ax.set_yticks([0]) # Hiding y ticks, as they don't really matter.

plt.savefig("Figures/Gaussian_Example.jpg")
plt.show()
```
